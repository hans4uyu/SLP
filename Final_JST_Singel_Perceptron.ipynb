```python
def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound


```

```python
outliers_kdr, lb_kdr, ub_kdr = identify_outliers(new_df, 'kdr')

print("Outliers KDR:\n", outliers_kdr)
print(f"KDR Lower Bound: {lb_kdr}, Upper Bound: {ub_kdr}")
```

```python
outliers_avg_dmg_delta, lb_avg_dmg_delta, ub_avg_dmg_delta = identify_outliers(new_df, 'avg_dmg_delta')

print("Outliers avg_dmg_delta:\n", outliers_avg_dmg_delta)
print(f"avg_dmg_delta Lower Bound: {lb_avg_dmg_delta}, Upper Bound: {ub_avg_dmg_delta}")
```

```python
outliers_acs, lb_acs, ub_acs = identify_outliers(new_df, 'acs')

print("Outliers ACS:\n", outliers_acs)
print(f"ACS Lower Bound: {lb_acs}, Upper Bound: {ub_acs}")
```

```python
import numpy as np

def log_transform(df, column):
    df[column] = df[column].apply(lambda x: np.log1p(x) if x >= 0 else -np.log1p(abs(x)))
    return df

new_df = log_transform(new_df, 'kdr')
new_df = log_transform(new_df, 'avg_dmg_delta')
new_df = log_transform(new_df, 'acs')

new_df.to_csv('log_transformed_valorant_match_clsfc.csv', index=False)
print("Data bersih dengan Log Transform berhasil disimpan sebagai 'log_transformed_valorant_match_clsfc.csv'")
```

```python
from sklearn.model_selection import train_test_split

encoded_df = pd.read_csv('one_hot_valorant_match_clsfc.csv')

X = encoded_df.drop(columns=['outcome_Win', 'outcome_Loss'])  # Asumsikan 'outcome' sudah di-encode
y = encoded_df[['outcome_Win', 'outcome_Loss']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)

X_train.to_csv('data_train_valorant_match_clsfc.csv', index=False)
X_test.to_csv('data_test_valorant_match_clsfc.csv', index=False)
y_train.to_csv('labels_train_valorant_match_clsfc.csv', index=False)
y_test.to_csv('labels_test_valorant_match_clsfc.csv', index=False)

print("Data latih dan uji berhasil disimpan sebagai 'data_train_valorant_match_clsfc.csv' dan 'data_test_valorant_match_clsfc.csv'")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
X_train = pd.read_csv('data_train_valorant_match_clsfc.csv')
X_test = pd.read_csv('data_test_valorant_match_clsfc.csv')
y_train = pd.read_csv('labels_train_valorant_match_clsfc.csv').iloc[:, 0]
y_test = pd.read_csv('labels_test_valorant_match_clsfc.csv').iloc[:, 0]
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Perceptron(max_iter=1000, random_state=123)
```

```python
from sklearn.linear_model import RidgeClassifier

model = RidgeClassifier(alpha=1.0)
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(model, X_train, y_train, cv=5)
print(f"Cross-Validation Scores: {cv_scores}")
print(f"Mean Cross-Validation Score: {cv_scores.mean()}")
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print(f"Akurasi model: {accuracy}")
print(f"Precision model: {precision}")
print(f"Recall model: {recall}")
print(f"F1-score model: {f1}")
```

```python
y_pred_train = model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_pred_train)
train_precision = precision_score(y_train, y_pred_train)
train_recall = recall_score(y_train, y_pred_train)
train_f1 = f1_score(y_train,y_pred_train)
```

```python
print(f"Akurasi pada data training: {train_accuracy}")
print(f"Precision pada data training: {train_precision}")
print(f"Recall pada data training: {train_recall}")
print(f"F1-score pada data training: {train_f1}")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Lose', 'Win'])
disp.plot()
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.show()
```
